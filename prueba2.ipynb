```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np
from statsmodels.graphics.mosaicplot import mosaic
from dateutil import parser
```

```python
df = pd.read_csv("Walmart.csv", sep=",")
columna_fecha = df['Date']

fechas_validas = []

for fecha in columna_fecha:
    try:
        fecha_parseada = parser.parse(fecha)
        fechas_validas.append(fecha_parseada)
    except ValueError:
        pass

if len(fechas_validas) > 0:
    print("Se encontraron fechas v√°lidas en la columna:")
    for fecha in fechas_validas:
        print(fecha)
else:
    print("No se encontraron fechas v√°lidas en la columna.")
```
