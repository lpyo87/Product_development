```python
from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt
import pandas as pd

# Supongamos que tienes dos variables categóricas 'variable1' y 'variable2' en tu DataFrame
data = {'variable1': ['A', 'B', 'A', 'B', 'A', 'B'],
        'variable2': ['X', 'X', 'Y', 'Y', 'Z', 'Z']}

df = pd.DataFrame(data)

# Crear un gráfico de mosaico
mosaic = mosaic(data=df, index=['variable1', 'variable2'])

# Dibujar el gráfico de mosaico
plt.figure(figsize=(8, 6))
mosaic.plot()
plt.title('Gráfico de Mosaico')

# Mostrar el gráfico
plt.show()
```
